```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("data.csv", index_col = 'Township')
index = np.arange(len(data.index))

plt.figure(figsize = (16,16)) # figure size

plt.subplot(2, 2, 1)
plt.plot(data.Male, label="Male")
plt.plot(data.Female, "r-.", label="Female")
plt.plot(data['Total(Male+Female)'], "g--", label="Total")
plt.title("the line graphs of population")
plt.legend()
plt.xlabel("Township")
plt.ylabel("Number of people")
plt.xticks(index, data.index, rotation = 60)

plt.subplot(2, 2, 2)
plt.plot(data.Immigration, "g--", label="Immigration")
plt.plot(data.Emigration, "r-o", label="Emigration")
plt.title("the line graphs of immigration and emigration")
plt.legend()
plt.xlabel("Township")
plt.ylabel("Number of people")
plt.xticks(index, data.index, rotation = 60)

plt.subplot(2, 2, 3)
plt.bar(data.index, data.Number_of_Households)
plt.title("the long bar graphs of household number")
plt.xlabel("Township")
plt.ylabel("Number of people")
plt.xticks(index, data.index, rotation = 60)

plt.subplot(2, 2, 4)
plt.plot(data.Birth, "g--", label="Birth")
plt.plot(data.Death, "r-o", label="Death")
plt.title("the line graphs of births and deaths")
plt.legend()
plt.xlabel("Township")
plt.ylabel("Number of people")
plt.xticks(index, data.index, rotation = 60)

plt.tight_layout()
plt.savefig('graphs.png', bbox_inches='tight')
plt.show()
```
