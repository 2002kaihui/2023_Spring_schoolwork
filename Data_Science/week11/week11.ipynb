```python
import pandas as pd
df = pd.read_csv('data.csv')

no_graduates = []
no_polUni = []
no_polCol = []
aver_mil = []
rate_polUni = []
rate_polCol = []
rate_mil = []

for count in range(6):
    if(count == 0):
        current = df[df["High school system"].isin(["National"]) & df["Gender"].isin(["female"])]
    elif(count == 1):
        current = df[df["High school system"].isin(["National"]) & df["Gender"].isin(["male"])]
    elif(count == 2):
        current = df[df["High school system"].isin(["county & city"]) & df["Gender"].isin(["female"])]
    elif(count == 3):
        current = df[df["High school system"].isin(["county & city"]) & df["Gender"].isin(["male"])]
    elif(count == 4):
        current = df[df["High school system"].isin(["private"]) & df["Gender"].isin(["female"])]
    elif(count == 5):
        current = df[df["High school system"].isin(["private"]) & df["Gender"].isin(["male"])]
    num_grad = current['Num of graduates'].sum()
    num_polu = current['Num of enroll in police university'].sum()
    num_polc = current['Num of enroll in police college'].sum()
    num_mil =  current['Num of enroll in military school'].sum()
    no_graduates.append(num_grad)
    no_polUni.append(num_polu)
    no_polCol.append(num_polc)
    aver_mil.append(round((num_mil / len(current)), 4))
    rate_polUni.append(round((num_polu / num_grad), 4))
    rate_polCol.append(round((num_polc / num_grad), 4))
    rate_mil.append(round((num_mil / num_grad), 4))

dists = {
    "High school system": ["National", "National", "county & city", "county & city", "private", "private"], 
    "Gender": ["female", "male", "female", "male", "female", "male"], 
    "Num of graduates": no_graduates, 
    "Num of enroll in police university": no_polUni, 
    "Num of enroll in police college": no_polCol,
    "Military school enrollment average": aver_mil, 
    "Police university enrollment rate": rate_polUni, 
    "Police college enrollment rate": rate_polCol,
    "Military school enrollment rate": rate_mil
}

df_dist = pd.DataFrame(dists)
df_dist
```
