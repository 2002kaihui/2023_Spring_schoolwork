```python
from pandas import DataFrame
import pandas as pd
from bs4 import BeautifulSoup
import requests
url = 'https://sys.ndhu.edu.tw/SA/XSL_ApplyRWD/ActApply.aspx?lang=en'
html = requests.get(url)
html.encoding="utf-8"

sp = BeautifulSoup(html.text, 'html.parser')

activity_name = []
certified_hour = []
registered_ppl = []
enroll_deadline = []
activity_timeslot = []
for i in range(10):
    idname = str('BodyContent_gvActs_lblGv_act_name_' + str(i))
    name = sp.find("span", id= idname).text
    activity_name.append(name)
    
    cerhour = str('BodyContent_gvActs_lblGv_xsl_check_' + str(i))
    hour = sp.find("span", id= cerhour).text.strip('Can / ')
    certified_hour.append(hour)
    
    regppl = str('BodyContent_gvActs_lblGv_reg_num_' + str(i))
    ppl = (sp.find("span", id= regppl).text).split()
    registered_ppl.append(ppl[0])
    
    enrdln = str('BodyContent_gvActs_lblGv_reg_dt_' + str(i))
    deadline = sp.find("span", id= enrdln).text
    enroll_deadline.append(deadline)
    
    acttime = str('BodyContent_gvActs_lblGv_act_dt_' + str(i))
    time = sp.find("span", id= acttime).text
    activity_timeslot.append(time)
    
data = {'Name of the activity': activity_name, 'Certificated': certified_hour, 'Number of people who have registered': registered_ppl,

        'Date and time of registeration': enroll_deadline, 'Date and time of activity': activity_timeslot}

df = pd.DataFrame(data, index=range(1, 11))
df
```
